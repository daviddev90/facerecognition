```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
num_of_data = 1
img_width_size = 300
img_height_size = 400
```

```python
testImage = np.zeros(num_of_data * img_width_size * img_height_size, dtype=np.int32).reshape(
    num_of_data, img_height_size, img_width_size
)
```

```python
img = np.array(Image.open('../Data/FaceGray/Aiden/image_0000.jpg'), dtype=np.int32)
testImage[0] = img
```

```python
plt.imshow(testImage[0], cmap='gray')
plt.show()
```

```python
dirNames = ['Aiden', 'Andrew', 'Cathy']
model2 = keras.models.load_model('../Data/CNN/best-gray-cnn-model.h5')
preds = model2.predict(testImage[0:1])
print(dirNames[np.argmax(preds[0])])
```
